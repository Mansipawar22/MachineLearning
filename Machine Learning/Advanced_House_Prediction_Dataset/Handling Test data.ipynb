## 1. Missing Values

```python
feature_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>1]
#it will give the list of feature with nan values and also the sum of nan values for each feature

for feature in feature_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(),4),'missing values')
```

```python
# Missing Values in Categorical features.

cat_features_na=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes=='O']

cat_features_na, len(cat_features_na)
```

```python
# To print the % of missing values of eac feature in cat_features_na

for feature in cat_features_na:
    print("{}: {}% missing values".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
dataset[cat_features_na].isnull().sum()
```

This shows the no. of nan values in each cat_features_na.

-> Now we will reaplace these null valyues with a term called 'Missing'.

```python
# we will generate a function first to replace the nan values
def replace_cat_features_na(dataset,cat_features_na):
    data=dataset.copy()
    data[cat_features_na]=data[cat_features_na].fillna('Missing')
    return data # it will return the updated dataset

dataset=replace_cat_features_na(dataset,cat_features_na)  # Calling the function
# Now lets check the no. of null values in the dataset

dataset[cat_features_na].isnull().sum()
```

As we can see there is no null value in any categorical feature.

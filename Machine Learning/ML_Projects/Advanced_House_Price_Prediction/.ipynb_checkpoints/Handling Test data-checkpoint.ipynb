### Missing Values in Categorical features

```python
cat_features_na=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes=='O']

cat_features_na, len(cat_features_na)
```

```python
# To print the % of missing values of eac feature in cat_features_na

for feature in cat_features_na:
    print("{}: {}% missing values".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
dataset[cat_features_na].isnull().sum()
```

This shows the no. of nan values in each cat_features_na.

-> Now we will reaplace these null valyues with a term called 'Missing'.

```python
# we will generate a function first to replace the nan values
def replace_cat_features_na(dataset,cat_features_na):
    data=dataset.copy()
    data[cat_features_na]=data[cat_features_na].fillna('Missing')
    return data # it will return the updated dataset

dataset=replace_cat_features_na(dataset,cat_features_na)  # Calling the function
# Now lets check the no. of null values in the dataset

dataset[cat_features_na].isnull().sum()
```

As we can see there is no null value in any categorical feature.

```python
dataset.head()
```

So we can see im Alley feature column all the nan values has been replaced with 'Missing'

### Missing values in Numerical Features

```python
numeric_features_na=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtype!='O']
numeric_features_na
```

There are 5 numerical features with null values. In trainning data there were only 3. Here 'BsmtFullBath' and '1bsmtHalfBath' are extra.

```python
dataset[numeric_features_na].isnull().sum()
```

```python
for feature in numeric_features_na:
    print('{}: {}% missing values'.format(feature,np.around(dataset[feature].isnull().mean(),4)))
```

```python
# Now I will fill these null values with median values

for feature in numeric_features_na:
    median_value = dataset[feature].median()
    # to create new columns to keep the check of missing values
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)
    
dataset[numeric_features_na].isnull().sum()    
```

So No null values

```python
dataset.head(50)
```

### Handling Rare Categorical Features

#### Rare Categorical Features
We have already replaced Rare categories from training datset we will replace the same categories from test data as well that's why we will use the same list.

These are the features which contains one or more category which comes only 1 or 2 time in the data means these categories are very rare so we can ignore these categories .

```python
categorical_features=[feature for feature in dataset.columns if dataset[feature].dtype=='O']
categorical_features,len(categorical_features)
```

```python
#To import the list of 'Rare_cate'

from csv import reader
# read csv file as a list of lists
with open('Rare_cate.csv', 'r') as read_obj:
    # pass the file object to reader() to get the reader object
    csv_reader = reader(read_obj)
    # Pass reader object to list() to get a list of lists
    Rare_cate_list = list(csv_reader)
    print(Rare_cate_list)
```

```python
for i in range(len(Rare_cate_list)):
    while('' in Rare_cate_list[i]) :
        Rare_cate_list[i].remove('')
Rare_cate_list
```

```python
categorical_features
```

```python
dataset[categorical_features].head(60)
```

Now we have list of categorical_features as well as nested list of rare_categories for all categorical_features.

```python
# To replace RARE CATEGORIES with 'Rare_var'

for feature in categorical_features:
    dataset[feature]=np.where(dataset[feature].isin(Rare_cate_list[categorical_features.index(feature)]),'Rare_var',dataset[feature])

dataset[categorical_features].head(50)
```

so all the Rare categories has been replaced by 'Rare_var'

```python
dataset[categorical_features].head(60)
```

```python
# Let's read the csv of labels_ordered[]
from csv import reader
# read csv file as a list of lists
with open('labels_ordered.csv', 'r') as read_obj:
    # pass the file object to reader() to get the reader object
    csv_reader = reader(read_obj)
    # Pass reader object to list() to get a list of lists
    labels_ordered = list(csv_reader)
    print(labels_ordered)
```

```python
for i in labels_ordered:
    while '' in i:
        i.remove('')

print(labels_ordered)
```

```python
for feature in categorical_features:
#     print(dataset[feature].unique())
    for category in dataset[feature].unique():
        if category not in labels_ordered[categorical_features.index(feature)]:
            labels_ordered[categorical_features.index(feature)].append(category)
            
labels_ordered
```

```python
for i in range(43):
    
    labels_ordered[i]={k:i for i,k in enumerate(labels_ordered[i],0)}
    print(labels_ordered[i])
    dataset[categorical_features[i]]=dataset[categorical_features[i]].map(labels_ordered[i])
```

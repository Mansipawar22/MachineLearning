# Univariate, Bivariate and Multivariate Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Iris.csv")
```

```python
df.head()
```

```python
df.shape
```

# Univariate Analysis

```python
df_setosa = df.loc[df['Species']=='Iris-setosa']    #this will give a new data set with setosa species only
df_virginica = df.loc[df['Species']=='Iris-virginica']
df_versicolor = df.loc[df['Species']=='Iris-versicolor']
```

```python
df_setosa.head()
```

```python
df_virginica.head()
```

```python
df_versicolor.head()
```

```python
plt.plot(df_setosa['SepalLengthCm'],np.zeros_like(df_setosa['SepalLengthCm']),'o')
plt.plot(df_virginica['SepalLengthCm'],np.zeros_like(df_virginica['SepalLengthCm']),'o')
plt.plot(df_versicolor['SepalLengthCm'],np.zeros_like(df_versicolor['SepalLengthCm']),'o')
plt.xlabel('Sepal Length')
plt.show
```

# Bivariant Analysis
    o Here we will use two features or two variables.
    o we will use a new function from seaborn known as FacetGrid()
    o hue=''  ask for the feature for which you are categorized date here i.e. "Species"

```python
sns.FacetGrid(df,hue='Species',height=5).map(plt.scatter,"SepalLengthCm","SepalWidthCm").add_legend();
plt.show()
```

# Multivariant Analysis
    o when we have more than two features to analyse
    o we will use pairplot() another new function of seanorn

```python
sns.pairplot(df,hue="Species",x_vars=["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"],y_vars=["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"],height=3)    #it should guve back total 16 graphs
plt.show()
```

# Correlation with Heatmap
A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection.

```python
df.corr()        #corr() for finding correlation
```

here we are getting only 3 features because all other fatures are of catigory type.
* it shows that if our total bill is increasing then tip will also increase as it has 67.% correlation
* between tip and size there is also positive correlation this means tip and size also directly proportion (48.9%) correlation.

# Joinplot
A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

## Univariate Analysis

```python
df.head()
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex',color='red')    #try to use diferrent parametres  instead of hist try to use hist, kid, reg
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg',color='g') 
```

# Pair plot
A pairs plot is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df, kind='reg', height=8)
```

pairplot() :- it will give multiple graphs of float and int type values. (total_ill, tip, size)

## KDE :- Kernel density estimation
kernel density estimation is a non-parametric way to estimate the probability density function of a random variable. Kernel density estimation is a fundamental data smoothing problem where inferences about the population are made, based on a finite data sample.

Note:-
* By default kde='True' so in y axis it will show in percentage
* when kde='False' it will give count

```python
sns.distplot(df['tip'],kde=False,bins=10)   # when kde is False in y axis it will give count
```

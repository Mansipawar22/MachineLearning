```python
x = np.arange(0, 4*np.pi,0.1)
y = np.sin(x)
plt.title("sine wave form")
           
#plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

```python
##subplots
#compute the x and y coordunates for points on sine and cosine curves 

x = np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x) 
y_cos = np.cos(x)

#set up a subplot grid that has height 2 and width 1,    [plt.subplot(2,1,1)] .  and set the first such subplot as active
plt.subplot(2,1,1)

#makinf of first plot
plt.plot(x,y_sin,'r--')
plt.title('Sine')

# Setting second subplot as active and making of sec plot
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g*-')
plt.title('Cosine')

#Showing the figure
plt.show()
```

# Bar Plot

```python
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y)
plt.bar(x2,y2,color = 'g')
plt.title('Bar Graph')
plt.ylabel('Y Axis')
plt.xlabel('X Axis')
plt.show()
```

# Box Plot using Matplotlib  - boxplot()

```python
# I am taking a normal distribution value from 0 to Standard deviation value wit step sizes 100 and the standard deviation will from 1 to 4

data = [np.random.normal(0, std, 100) for std in range(1,4)]
```

```python
data     #this will print an array coz data is an array
```

```python
#rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
```
